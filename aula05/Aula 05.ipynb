```python
#Calcular a área de um quadrilátero, com dois lados iguais. Para isso faz a leitura de dois números inteiros, representando os lados e imprime os valores lidos e a área calculada.base, altura, area representam os nomes das variáveis – podem ser do tipo inteiro ou do tipo real.

base = int(input("Insira a base em metros: "))
altura = int(input("Insira a altura em metros: "))
area = base * altura

print("A area do quadrilatero é: ", area)
```
