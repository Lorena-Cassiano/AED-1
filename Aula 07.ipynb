```python
#A impressão deve ocorrer apenas com números IMPARES. Quando chegar a 10 o loop deve ser interrompido.

n = 0

while n <= 10:
    if n % 2 == 0:
        n += 1
        continue
    
    print(n)

    if n == 10:
        break

    n += 1
```

```python
#2- Faça um algoritmo que calcule a soma dos primeiros 50 números pares. Este algoritmo não recebe valores do teclado. 
# WHILE

n = 0

while n <= 50:
    if n % 2 != 0:
        n += 1
        continue

    print(n)
    n += 1
```

```python
#2- Faça um algoritmo que calcule a soma dos primeiros 50 números pares. Este algoritmo não recebe valores do teclado. 
# FOR

n = 0 

for n in range(1,50):
    if n % 2 != 0:
          n+=1
          continue

    print (n)
```
